**IMPORTING LIBRARIES**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**LOADING THE DATASET**

```python
data = pd.read_csv("student-mat.csv",sep=";")
```

**Data Cleaning**

```python
data.drop_duplicates()
```

**Data Analysis Questions**

 ***1. What is the average score in math (G3)?***

```python
average_score=data["G3"].mean()
print("The average_score in maths:",average_score)
```

***2. How many students scored above 15 in their final grade (G3)?***

```python
score=len(data[data['G3']>15])
print("students scored above 15 in their final grade (G3):",score)
```

***3. Is there a correlation between study time (study time) and the final grade (G3)?***

```python
corr=data["studytime"].corr(data["G3"])
print("correlation between study time (study time) and the final grade (G3)",corr.round(2))
```

 ***4. Which gender has a higher average final grade (G3)?***

```python
average_gender=data.groupby("sex")["G3"].mean()
print(average_gender)
```

**DATA VISUALIZATION**

**1. Plot a histogram of final grades (G3)**

```python
plt.figure(figsize=(8,6))
plt.hist(data['G3'],color="orange")
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")
plt.title("Distribution of Final Grades")
plt.show()
```

**2. Create a scatter plot between study time (study time) and final grade (G3)**

```python
plt.figure(figsize=(4,8))
sns.scatterplot(data=data,x="studytime",y="G3",color="green")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.title("Relationship between Study Time and Final Grade")
plt.show()
```

**3. Create a bar chart comparing the average scores of male and female students**

```python
#bar chart
plt.figure(figsize=(4,5))
average_gender.plot(kind='bar',color=["red","blue"])
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.title("Average Scores of Male and Female Students")
plt.show()

```
